## Load Dataset

Download dataset dari COCO 17 melalui fifty one dengan menspesifikasikan hanya Class 'Person'

```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as four
import fiftyone.utils.yolo as fouy
```

```python
dataset = foz.load_zoo_dataset(
    "coco-2017",
    # splits=["train","validation", "test"],
    label_types=["detections"],
    classes=["person"],
    max_samples=1000,
)
```

```python
dataset.count()
```

launch DataView

```python
session = fo.launch_app(dataset)
```

```python
from fiftyone import ViewField as F

# Find samples that have a "Vehicle registration plate"
person_labels_only_ds = dataset.filter_labels("ground_truth", F("label")=="person")
```

```python
person_labels_only_ds.count()
```

```python
nsample = person_labels_only_ds.count()
ntrain = int(nsample * 0.7)
ntest = int(nsample * 0.2)
nval = nsample - (ntrain + ntest)

shuffled_view = person_labels_only_ds.shuffle()

train_view = shuffled_view[:ntrain]
test_view = shuffled_view[ntrain:ntrain+ntest]
val_view = shuffled_view[-nval:]

for sample in train_view.iter_samples(autosave = True):
    sample.tags.append("train")

for sample in test_view.iter_samples(autosave = True):
    sample.tags.append("test")

for sample in val_view.iter_samples(autosave = True):
    sample.tags.append("val")
```

```python
print(train_view.count())
print(test_view.count())
print(val_view.count())
```
